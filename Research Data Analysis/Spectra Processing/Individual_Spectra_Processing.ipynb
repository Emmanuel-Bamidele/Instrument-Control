```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.signal import savgol_filter

directory = "s1-1A_per_s00833.txt"
column_names = ['Wavelength', 'Intensity']
df = pd.read_csv(directory, delimiter="\t", names=column_names, header=None)
#df.head()

plt.rc('font', family='serif', serif='Times')
plt.rc('xtick', labelsize=22)
plt.rc('ytick', labelsize=22)
plt.rc('axes', labelsize=22)

plt.rcParams['font.size'] = 18
plt.rcParams['axes.linewidth'] = 2

# width as measured in inkscape
width = 13
height = width / 1.5

fig, ax = plt.subplots()
fig.subplots_adjust(left=.15, bottom=.16, right=.99, top=.97)

x = df.Wavelength
y = df.Intensity
y2 = savgol_filter(y, 301, 3)
#plt.plot(x, y/1000)
plt.plot(x, y2/1000, color='red', lw=4)

ax.set_xlabel('Wavelength (nm)', labelpad=18, fontsize=22)
ax.set_ylabel('Intensity (a.u)', labelpad=18, fontsize=22)
ax.set_ylim(-0.1, (df.Intensity.max() + 500)/1000)

fig.set_size_inches(width, height)
plt.savefig('plot.jpeg', dpi=600, transparent=False, bbox_inches='tight')
plt.show()
```
